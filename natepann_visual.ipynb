```python
from openpyxl import load_workbook
```

```python
read_wb = load_workbook("./natepann.xlsx")
read_ws = read_wb.active

list_excel = []

for i in range(1,101):
    result = read_ws.cell(i,1).value
    list_excel.append(read_ws.cell(i,1).value)
```

```python
list_excel
```

```python
from konlpy.tag import Kkma
import collections
```

```python
kkma = Kkma()
list_temp = []
```

```python
for row in list_excel :
    nouns_for_row = kkma.nouns(row)
    list_temp = list_temp + nouns_for_row
```

```python
list_temp
```

```python
result_list = []
```

```python
for check in list_temp:
    if len(check) > 1:
        result_list.append(check)
        
print(result_list)
```

```python
last_string = ""

for i in result_list:
    last_string += i + " "
```
